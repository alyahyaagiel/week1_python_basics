## Day 1 – Python Basics & Setup

```

# Print & Input
print("Hello, Data Science World!")

nama = input("Masukkan nama anda: ")
umur = int(input("Masukkan umur anda: "))
print(f"Hai {nama}, umur anda {umur} tahun.")

# Data types
x = 10
y = 3.5
z = "Data Science"
status = True

print(type(x), type(y), type(z), type(status))

# Operator
a = 10
b = 5
print(a + b, a - b, a * b, a / b)
print(a > b, a == b, a != b)
```

## Day 2 – Conditions

```

umur = int(input("Masukkan umur anda: "))

if umur >= 18:
    print("Anda Dewasa.")
elif umur >= 13:
    print("Anda Remaja.")
else:
    print("Anda Kanak-kanak.")

markah = int(input("Masukkan markah anda: "))

if markah >= 80:
    print("Gred A")
elif markah >= 60:
    print("Gred B")
else:
    print("Gred C atau bawah")
```

## Day 3 – Loops

```

# For loop
for i in range(1, 6):
    print(f"Nombor: {i}")

# While loop
count = 1
while count <= 5:
    print(f"Kiraan: {count}")
    count += 1

# Break & continue
for i in range(10):
    if i == 5:
        break
    print(i)
```

## Day 4 – Data Structures

```

buah = ["epal", "mangga", "pisang"]
buah.append("anggur")
print(buah)

pelajar = {"Ali": 85, "Siti": 92, "Ravi": 75}
print(pelajar["Siti"])

for nama, markah in pelajar.items():
    print(f"{nama}: {markah}")
```

## Day 5 – String & Function

```

teks = "data science is fun"
print(teks.upper())
print(teks.replace("fun", "awesome"))

def kira_bmi(berat, tinggi):
    bmi = berat / (tinggi ** 2)
    return round(bmi, 2)

print(kira_bmi(60, 1.7))
```

## Day 6 – Mini Project: Student Grade System

```

def tentukan_gred(markah):
    if markah >= 80:
        return "A"
    elif markah >= 65:
        return "B"
    elif markah >= 50:
        return "C"
    else:
        return "D"

pelajar = {}
jumlah = int(input("Berapa orang pelajar? "))

for i in range(jumlah):
    nama = input("Nama pelajar: ")
    markah = int(input("Markah: "))
    pelajar[nama] = tentukan_gred(markah)

print("\nSenarai Pelajar & Gred:")
for n, g in pelajar.items():
    print(f"{n}: {g}")
```
